Operadores Aritméticos

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)

print(a / b)  # Divisão Comum 10 / 3 = 3.33333
print(a // b) # retorna apenas a parte inteira do resultado 10 / 3 = 3
print(a % b)  # retorna o 'resto' da divisão 10 / 3 = 3 * 3 = 9 , resta 1

print(a ** b) # potencia 10³
```

Estrutura condicional IF

```python
# Condição Simples - Status por avaliação

avaliacao = 4.2

if avaliacao >= 7.0:
    print("Legal, obrigada por sua avaliação!")

else: 
    print("Que pena! Vamos melhorar...")
```

```python
# Condição Composto

avaliacao = 5.5

if avaliacao >= 7.0:
    print("Legal, obrigada por sua avaliação!")

elif avaliacao >= 5.0:
    print("Legal, faremos alguns ajustes!")

else: 
    print("Que pena! Vamos melhorar...")
```

```python
# Condição aninhada

avaliacao = 5.5

if avaliacao >= 7.0:
    print("Legal, obrigada por sua avaliação!")

elif avaliacao >= 5.0:
    resposta = input("Quer justificar sua respostas? [S ou N]: ").upper()

    if resposta == "S":
        print("Ok, então justifique: ")
    elif resposta == 'N':
        print("Ok.. ")
    else:
        print("Resposta Inválida!")

else: 
    print("Que pena! Vamos melhorar...")
```

Match/Case

```python
menu = int(input("1 Cadastrar | 2 Visualizar | Opção: "))

match menu:
    case 1:
        print("escolheu 1")
    
    case 2:
        print("escolheu 2")

    case _:
        print("Resposta Inválida!")
```
