```python
# Entrada de dados
nome_funcionario = str(input("Nome Funcion√°rio: ")).title()
cargo = int(input("1 - Analista Financeiro| 2 - Assistente Financeiro| 3 - Coordenador"))
salario = float(input("Sal√°rio: "))
anos_experiencia = int(input("Anos de Experi√™ncia: "))
print(nome_funcionario)
print(cargo)
print(salario)
print(anos_experiencia)
```

```python
# Identificar o cargo escolhido

match cargo:
    case 1:
        cargo = "Analista Financeiro"
    case 2: 
        cargo = "Assistente Financeiro"
    case 3: 
        cargo = "Coordenador"

    case _:
        print("Resposta Inv√°lida!")

cargo

    
```

```python
# Classifica√ß√£o do funcion√°rio por tempo de experi√™ncia

if anos_experiencia <= 2 and anos_experiencia > 0:
    senioridade = "J√∫nior"
elif anos_experiencia >= 3 and anos_experiencia <= 5:
    senioridade = "Pleno"
elif anos_experiencia >= 6 and anos_experiencia <= 40:
    senioridade = "S√™nior"

else:
    senioridade = "Erro"

```
