Importando bibliotecas

```python
from datetime import datetime
import json
```

Carregando arquivo

```python
caminho = 'dados\contas.json'

try:
    with open(caminho, 'r', encoding="UTF8") as arquivo:
        contas = json.load(arquivo)

except FileNotFoundError as e:
    print(e)
```

Tratamento de Dados - Convertendo Datas

```python
# Convertendo a data texto em data real
for conta in contas:
    conta['vencimento'] = datetime.strptime(conta['vencimento'], "%d/%m/%Y")

# Verificando a conversÃ£o
contas[0]['vencimento']
```

Contas Vencidas

```python
contas_vencidas = []
contas_vencidas_pagar = []
contas_vencidas_receber = []
contas_erro_tipo = []

for conta in contas:
    if conta['vencimento'] < datetime.today():
        contas_vencidas.append(conta)

        if conta['tipo'] == 'Pagar':
            contas_vencidas_pagar.append(conta)

        elif conta['tipo'] == 'Receber':
            contas_vencidas_receber.append(conta)
        
        else:
            contas_erro_tipo.append(conta)

print(len(contas_vencidas))
print(len(contas_vencidas_pagar))
print(len(contas_vencidas_receber))
print(len(contas_erro_tipo))
```

```python
saldo_vencido = 0

for conta in contas:
    if conta['vencimento'] < datetime.today() and conta['tipo'] == 'Pagar'and conta['status'] == 'Em aberto':
        saldo_vencido += conta['valor']

print(f"Saldo de Contas Ã  Pagar Vencidas: R$ {saldo_vencido: ,.2f}")
```

Contas Ã  Pagar

```python
saldo_a_pagar = 0

for conta in contas:
    if conta['tipo'] == 'Pagar':
        saldo_a_pagar += conta['valor']

print(f"Total de Contas Ã  Pagar: R$ {saldo_a_pagar: ,.2f}")
```

Contas Ã  Receber

```python
saldo_a_receber = 0

for conta in contas:
    if conta['tipo'] == 'Receber':
        saldo_a_receber += conta['valor']

print(f"Total de Contas Ã  Receber: R$ {saldo_a_receber: ,.2f}")
```

Resultado Consolidado

```python
#contas_pagas = saldo_a_pagar - saldo_vencido

#saldo_final1 = saldo_a_receber - contas_pagas 
#saldo_final2 = saldo_a_receber - contas_pagas - saldo_vencido

saldo_final = saldo_a_receber - saldo_a_pagar

#print(f'Saldo antes do pagamento das contas vencidas: R$ {saldo_final1: ,.2f}')
#print(f'O saldo final apÃ³s pagamento de contas vencidas: R$ {saldo_final2: ,.2f}')
print(f'O saldo final Ã©: R$ {saldo_final: ,.2f}')
```
