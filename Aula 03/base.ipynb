```python
# Com parâmetros e argumentos

# Parametro: na construção da função
# argumento: na hora de chamar a função

def dizoi(nome): # "nome" seria o parametro
    return f'Oi, {nome}' 

#dizoi("Monique!") # "Monique" seria o argumento


def cantar_parabens(nome):
    return f'Oi, {nome}! Parabéns pra você, nessa data querida!'

#cantar_parabens(input("Aniversariante: "))

def soma2valores(num1, num2):
    return num1 + num2

print(soma2valores(num1=33, num2=69))
soma2valores(55,33)
```

```python
def geradorDeDesconto(valor, desconto = 0.1, *args, **kwargs): # o desconto imposto aqui, só vai ser usado se não for argumentado um valor na hr de chamar a função
    # o parametro *args aceita a diversos argumentos mas não pode ser nomeado ex "for _ in *args| o **kwargs recebe quantidade flexivel de argumentos mas ele pode ser nomeado ex "for c in cidade"
    """
    essa função serve para gerar desconto
    Explicar:
    Valor: Pode ser float ou int
    Desconto: precisa ser informado decimal
    """
    return valor * desconto

geradorDeDesconto(200, 0.2)
```

```python
import json

try:
    with open(r"dados\contas.json", 'r', encoding="UTF8") as arquivo:
        contas = json.load(arquivo)

except FileNotFoundError as e:
    print(e)
```

```python
# Converter data em texto para data real (datetime)
from datetime import datetime

for dados in contas:
    dados['vencimento'] = datetime.strptime(dados['vencimento'], "%d/%m/%Y")
# strf converte 
# strp converte e deixa explicito como quer a saída

```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/>
        Análise (Vencimentos)
    </strong>
</p>

```python
# Quantas contas estão vencidas?

from datetime import datetime

hoje = datetime.now()
contas_vencidas = []

for dados in contas:
    if dados['vencimento'] < hoje and dados['status'] == 'Em aberto' and dados['tipo'] == 'Pagar': 
        contas_vencidas.append(dados)

print(contas_vencidas)
len(contas_vencidas)
    
```

```python
# Total em contas vencidas?
valor_vencido = 0

for dados in contas_vencidas:
    valor_vencido += dados['valor']

print(f"Total Vencido: R$ {valor_vencido: ,.2f}")
```
