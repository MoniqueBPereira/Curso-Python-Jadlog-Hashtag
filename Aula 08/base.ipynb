Importando Lib's

```python
import pandas as pd
```

Criando dataframe

```python
df_entregas = pd.DataFrame({
    'id_entrega' : [1,2,3], # chave prim√°ria
    'filial'     : ['RJ','MG','SP'],
    'status'     : ['ENTREGUE','ATRASADO','ENTREGUE']
})

df_custos = pd.DataFrame({
    'id_custo'   : [1,2,3],
    'id_entrega' : [1,1,3],
    'custo'      : [150.89,147.56,278.26]
})
```

JOIN

```python
# Inner Join 

pd.merge(
    df_entregas, df_custos,
    on='id_entrega',
    how='inner'
)
```

```python
# Left Join 

pd.merge(
    df_entregas, df_custos,
    on='id_entrega',
    how='left'
)
```

```python
# Full Outer Join 

pd.merge(
    df_entregas, df_custos,
    on='id_entrega',
    how='outer'
)
```

```python
df_entregas_idx = df_entregas.set_index('id_entrega')
df_custos_idx = df_custos.set_index('id_entrega')

df_join = df_entregas_idx.join(df_custos_idx, how='left')

df_join.reset_index()
```
