```python
entrada = []
saida = []
valor_entrada = 0
valor_saida = 0
historico_entrada = []
historico_saida = []
```

```python
# Menu
 
while True:
    menu = int(input("1.Dep√≥sito | 2.Saque | 3.Sair | Op√ß√£o:"))
 
    match menu:
            case 1:
                entrada = float(input("Digite o valor de Dep√≥sito: "))
                valor_entrada += entrada
                historico_entrada.append(entrada)
         
            case 2:
                saida = float(input("Digite o valor de saque: "))
                if saida < entrada:
                    valor_saida += saida
                    historico_saida.append(saida)
                    
                else:
                     print('Sem saldo!')
            
            case 3:
                break
 
            case _:
                print("Op√ß√£o Inv√°lida")
 
valor_final = valor_entrada - valor_saida
print(f'O valor dispon√≠vel na conta √© R$ {valor_final}\nHist√≥rico de entrada: {historico_entrada}\nHist√≥rico de saida √©: R$ {historico_saida}')
```

```python
saldo = 0
lista_transacoes = []

while True:
    menu = int(input("1.Dep√≥sito | 2.Saque | 3.Sair | Op√ß√£o:"))
 
    if menu in (1,2):
        valor = float(input('Valor: '))

    match menu:
            case 1:
                print(f'Dep√≥sito Conclu√≠do. | R$ {valor}')    
                saldo += valor
                lista_transacoes.append(f"DEP: {valor}")

            case 2:
                if valor < saldo:
                    saldo -= valor
                    print(f'Saque Conclu√≠do no valor de R$ {valor} | Saldo Atual: R$ {saldo}')    
                    lista_transacoes.append(f"SAQ: {valor * -1}")
                else:
                     print(f"Saldo Insuficiente!")
            case 3:
                break
 
            case _:
                print("Op√ß√£o Inv√°lida..")
 
print(lista_transacoes)
print(f"Saldo: R$ {saldo}")
```
