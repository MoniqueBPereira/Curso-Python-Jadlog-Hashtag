```python
# repetindo um texto 5 vezes com for

for contador in range(5): print("Monique")

# repetindo um texto 5 vezes com while

numero = 0
while numero <= 5:
    print("Batista")
    numero += 1
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong></br></br></br></br>
        Validando um senha de brincadeira
    </strong>
</p>

```python
senha_cadastrada = input('Senha: ')
senha_entrada = ''
tentativas = 0

while senha_cadastrada != senha_entrada:
    senha_entrada = input('Digite sua senha: ')
    tentativas += 1

    if tentativas == 3 and senha_cadastrada != senha_entrada:
        print("Tentativas máximas atingida")
        break

if senha_cadastrada == senha_entrada:
    print('Acesso permitido')
```

        While com Break
    </strong>
</p>

```python
while True:
    menu = int(input('1_Cadastrar | 2_Visualizar | 3_Sair | Opção: '))

    match menu:
        case 1:
            nome_franquia = input('Franquia: ')
        case 2:
            print(nome_franquia)
        case 3:
            break
        case _:
            print('Opção inválida!')
```

        # Exemplo: Imagine que cada franquiado me pague 10% de seu faturamento mensal 
    </strong>
</p>

```python
# Entrada de valores

franquias = ['LJ Rio de Janeiro 55', 'LJ São Paulo 23', 'PA Rio 50', 'LJ Sao 88']

pg_franquias = []

for franquia in franquias:
    valor = float(input(f'{franquia} | Valor: '))
    
    pg_franquias.append(franquia)
    pg_franquias.append(valor)

pg_franquias
```

```python
# Cálculo
soma = 0
for indice, dado in enumerate(pg_franquias):
    if indice % 2 == 1:
        soma += dado * 0.1

print(round(soma,2))
```
